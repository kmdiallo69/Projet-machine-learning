#### Credit card Fraud Detection

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
## loading dataset to a pandas DataFrame
ccard=pd.read_csv('/Users/km.diallo/Documents/gold_project/creditcard.csv')
```

```python
# dataset informations
ccard.info()
```

```python
# checking the number of missing values in each column
ccard.isnull().sum()
```

```python
# distribution of legit transactions & fraudulent transactions
ccard['Class'].value_counts()
```

 This Dataset is highly unbalanced:<br>
    0 -> Normal Transaction<br>
    1 -> Fraud Trnasaction

```python
# separating the data for analysis
legit=ccard[ccard.Class==0]
fraud=ccard[ccard.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
ccard.groupby('Class').mean()
```

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions<br>
Number of Fraudulent Transactions --> 492

```python
legit_sampl=legit.sample(n=492)
```

```python
## Concatenating two DataFrames
new_dataset = pd.concat([legit_sampl,fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

 Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model=LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
